## Matplotlib example workflow

```python
# 0. import matplotlib and get it ready for plotting in Jupter
%matplotlib inline
import matplotlib.pyplot as plt

#1. Prepare data
x = [1,2,3,4]
y = [11,22,33,44]

# 2.setup plot
fig, ax = plt.subplots(figsize=(5,10)) # figsize = hight , width

# 3. Plot data
ax.plot(x,y)

# 4. Customize plot
ax.set(title="Simple Plot",
       xlabel="x-axis",
       ylabel="y-axis")

# 5. Save & show (you save the whole figure )
fig.savefig("Images\sample-plot.png")
```

## Making figures with Numpy arrays
We eant:
* Line plot
* Scatter plot
* Histogram
* Subplots

```python
import numpy as np
```

```python
# Create some data
x = np.linspace(0,10,100)
x[:20]
```

```python
# plot the data and create a line plot
fig,ax = plt.subplots()
ax.plot(x, x**2);
```

```python
# Use same data to make a scatter
fig,ax = plt.subplots()
ax.scatter(x, np.exp(x))
```

```python
# Another Scatterplot
fig,ax = plt.subplots()
ax.scatter(x, np.sin(x))
```

```python
# Make a plot from dictionary using bar chart
nut_butter_price = {"Almond butter":10,
                   "Peanut butter":8,
                   "Cashew butter":12}
fig,ax = plt.subplots()
ax.bar(nut_butter_price.keys(),nut_butter_price.values())
ax.set(title ="Dan's Nut Butter Stores",
      ylabel="Price ($)");
```

```python
fig,ax = plt.subplots()
ax.barh(nut_butter_price.keys(),nut_butter_price.values());
```

```python
#Make some data for histograms and plot it
x = np.random.randn(1000)
fig,ax = plt.subplots()
ax.hist(x);
```

## Two options for subplots

```python
## Subplot option 1
fig,((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,
                                         ncols=2,
                                         figsize=(10,5))
#plot to each different axis
ax1.plot(x,x/2);
ax2.scatter(np.random.random(10),np.random.random(10));
ax3.bar(nut_butter_price.keys(), nut_butter_price.values());
ax4.hist(np.random.randn(1000));
```

```python
#Subplot option 2
fig,ax = plt.subplots(nrows=2,
                     ncols=2,
                     figsize=(10,5))

#Plot to each different index
ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(10),np.random.random(10));
ax[1,0].bar(nut_butter_price.keys(), nut_butter_price.values());
ax[1,1].hist(np.random.randn(1000));
```
