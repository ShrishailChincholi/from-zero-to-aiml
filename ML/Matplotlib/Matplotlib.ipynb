## Matplotlib example workflow

```python
# 0. import matplotlib and get it ready for plotting in Jupter
%matplotlib inline
import matplotlib.pyplot as plt

#1. Prepare data
x = [1,2,3,4]
y = [11,22,33,44]

# 2.setup plot
fig, ax = plt.subplots(figsize=(5,10)) # figsize = hight , width

# 3. Plot data
ax.plot(x,y)

# 4. Customize plot
ax.set(title="Simple Plot",
       xlabel="x-axis",
       ylabel="y-axis")

# 5. Save & show (you save the whole figure )
fig.savefig("Images\sample-plot.png")
```

## Two options for subplots

```python
## Subplot option 1
fig,((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,
                                         ncols=2,
                                         figsize=(10,5))
#plot to each different axis
ax1.plot(x,x/2);
ax2.scatter(np.random.random(10),np.random.random(10));
ax3.bar(nut_butter_price.keys(), nut_butter_price.values());
ax4.hist(np.random.randn(1000));
```

```python
#Subplot option 2
fig,ax = plt.subplots(nrows=2,
                     ncols=2,
                     figsize=(10,5))

#Plot to each different index
ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(10),np.random.random(10));
ax[1,0].bar(nut_butter_price.keys(), nut_butter_price.values());
ax[1,1].hist(np.random.randn(1000));
```

## Which one should you use? (pyplot vs matplotlib OO method?)
* When plotting somethig quickly,okay to use the pyplot method
* When plotting something more advanced , use the OO method

```python
heart_disease.head()
```

```python
over_50 =heart_disease[heart_disease["age"] >50]
over_50.head()
    
```

```python
#Pyplot method
over_50.plot(kind='scatter',
          x='age',
          y='chol',
          c='target');
```

```python
##OO method
fix , ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target',
             ax=ax);

# ax.set_xlim([45,100]);
```

```python
fix , ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target',
             ax=ax);

ax.set_xlim([45,100]);
```

```python
# OO method from scratch
fix, ax = plt.subplots(figsize=(10,6))

# plot the data
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"]);

```

```python
# OO method from scratch
fix, ax = plt.subplots(figsize=(10,6))

# plot the data
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"]);
# Customize the plot
ax.set(title="Heart Disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol")

#Add a legend
ax.legend(*scatter.legend_elements(), title="Target");

#Add horizontal line
ax.axhline(over_50["chol"].mean(),
           linestyle='--');
```

```python
over_50.head()
```

```python
# Subplot of chol,age,thalach
fig,(ax0, ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10,10),
                             sharex=True)

#Add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"]);

#customize ax0
ax0.set(title="Heart Disease and Cholesterol Levels",
        # xlabel="Age",
        ylabel="Cholesterol")
#Add a legend to ax0
ax0.legend(*scatter.legend_elements(),title="Target");

# Add a meanline
ax0.axhline(y=over_50["chol"].mean(),
            linestyle="--");

#Add data to ax1
scatter = ax1.scatter(x=over_50["age"],
                      y=over_50["thalach"],
                      c=over_50["target"])

#customize ax1
ax1.set(title="Heart Disease and Max Heart Rate",
        xlabel="Age",
        ylabel="Max Heart Rate");

#Add a legend to ax1
ax1.legend(*scatter.legend_elements(),title="Target");

# Add a meanline
ax1.axhline(y=over_50["thalach"].mean(),
            linestyle="--");

#add a title to the figure
fig.suptitle("Heart Disease Analysis", fontweight="bold");
```
