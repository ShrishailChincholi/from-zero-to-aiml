```python
# Standard deviation = a measureof how spread out a group of number is form the mean
np.std(a2)
```

```python
# Variance = measure of the average degree to which each number is difference to the mean
# Higher  Variance = wider range of number
# Lower Variance = lower range of number
np.var(a2)
```

```python
# Standard deviation = squareroot of variance 
np.sqrt(np.var(a2))
```

```python
# Demo of std and var
high_var_array = np.array([1, 100, 200, 300, 4000, 5000])
lower_var_array = np.array([2, 4, 6, 8, 10])
```

```python
np.var(high_var_array), np.var(lower_var_array)
```

```python
np.std(high_var_array) , np.std(lower_var_array)
```

```python
np.mean(high_var_array)  , np.mean(lower_var_array)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.hist(high_var_array)
plt.show()
```
