## Dot product exmples( nut butter sales)

```python
np.random.seed(0)
#Nuumber of jars sold
sales_amount = np.random.randint(20, size = (5,3))
sales_amount
```

```python
#Create weekly_sales DataFrame
weekly_sales = pd.DataFrame(sales_amount , index=["Mon", "Tues", "Web", "Thurs", "Fri"],
                           columns=["Almond butter", "Peanbutter", "Cashew butter"])
weekly_sales
```

```python
#Create Prices array
prices = np.array([10, 8, 12])
prices
```

```python
#Create butter_prices DataFrame
butter_prices = pd.DataFrame(prices.reshape(1,3), index=["Prices"],columns=["Almond butter", "Peanbutter", "Cashew butter"])
butter_prices
```

```python
total_shape = prices.dot(sales_amount)
```

```python
prices.shape
```

```python
sales_amount.shape
```

```python
#Shape aren't aligned, let's transpose
total_sales = prices.dot(sales_amount.T)
total_sales
```

```python
# create daily_sales
butter_prices.shape , weekly_sales.shape
```

```python
daily_sales = butter_prices.dot(weekly_sales.T)
daily_sales
```

```python
weekly_sales
```

```python
#Doesn't work,  not right shape
weekly_sales["Total ($) "] = daily_sales
weekly_sales
```

```python
weekly_sales["Total ($)"] = daily_sales.T
weekly_sales
```
