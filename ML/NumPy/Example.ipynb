```python
import numpy as np
```

```python
np.random.seed(0)
student_marks = np.random.randint(100,size = (4,6))
student_marks
```

```python
student = np.DataFrame(student_marks , index =["Jhon", "Alex", "Joy", "Robb"], colums=["Maths", "Scinece", "Biology", "Physices", "Computer Scinece", "Chemistry"])
student
```

```python
import pandas as pd
```

```python
student = pd.DataFrame(student_marks , index =["Jhon", "Alex", "Joy", "Robb"], columns=["Maths", "Scinece", "Biology", "Physices", "Computer Scinece", "Chemistry"])
student
```

```python
weights = [1, 1, 1, 1, 1, 1]
```

```python
student["Total_scroe"] = student.dot(weights)
```

Car Performance Score using DOT PRODUCT

```python
np.random.seed(0)
car_data = np.random.randint(20,size=(4,6))
car_data
```

```python
columns = [
    "Mileage_kmpl",
    "Engine_HP",
    "Safety_Rating",
    "Comfort_Rating",
    "Maintenance_Cost",
    "Resale_Value"
]
```

```python
car = pd.DataFrame(car_data, index = ["car_A", "car_B", "car_C", "car_D"],
                  columns = columns)
car
```

```python
range = [
    0.20,   # Mileage
    0.25,   # Engine power
    0.20,   # Safety
    0.15,   # Comfort
   -0.10,   # Maintenance cost (negative impact)
    0.10    # Resale value
]
range
```

```python
car["Performance_Score"] = car.dot(range)
```

```python
car
```
